#  Радиальное распределение волновой функции

* 1s

$$
\rho_{1s}(r) = 4r^2\exp(-2r)
$$

* 2s

$$
\rho_{2s}(r) = \frac{1}{2}r^2\left(1-\frac{r}{2}\right)^2\exp(-r)
$$

* 3s

$$
\rho_{3s}(r) = \frac{4}{27}r^2\left(1-\frac{2r}{3}+\frac{2r^2}{27}\right)^2\exp\left(-\frac{2r}{3}\right)
$$

* 2p

$$
\rho_{2p}(r) = \frac{1}{24}r^4\exp(-r)
$$

```python
from numpy import exp, linspace
```

```python
r = linspace(0, 30, 3000)
```

```python
rho = {
    '1s': 4*r**2*exp(-2*r),
    '2s': 0.5*r**2*(1-r/2)**2*exp(-r),
    '3s': 4/27*r**2*(1-2*r/3+2*r**2/27)**2*exp(-2*r/3),
    '2p': 1/24*r**4*exp(-r)
}
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use(['default', 'seaborn-paper', 'seaborn-whitegrid', 'seaborn-deep'])
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

for k in rho:
    ax.plot(r, rho[k], label=k)

ax.set_xlim(0, 30)
ax.set_ylim(0, 0.55)

ax.legend()

plt.xlabel('$r/r_B$')
plt.ylabel('$\\rho(r)$')

plt.savefig('/home/artem/projects/github/atomic-physics/img/hydrogen-radial-functions.svg')
#plt.show()
```
